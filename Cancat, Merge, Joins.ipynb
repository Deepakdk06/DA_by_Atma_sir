```python
import pandas as pd
```

```python
dummy_data = {
            'id':['1','2','3','4','5'],
            'Feature1':['A','C','E','G','I'],
            'Feature2':['B','D','F','H','J']
}
df=pd.DataFrame(dummy_data)
```

```python
df
```

```python
dummy_data1 = {
            'id':['1','2','6','7','8'],
            'Feature1':['K','M','O','Q','S'],
            'Feature2':['L','N','P','R','T']
}
df1=pd.DataFrame(dummy_data1)
```

```python
df1
```

```python
dummy_data2 = {
            'id':['1','2','3','4','5','7','8','9','10','11'],
            'Feature2':[12,13,14,15,16,17,15,12,13,23]
}
df2=pd.DataFrame(dummy_data2)
```

```python
df2
```

# Concat

```python
data_concat=pd.concat([df,df1])   # 'df' will shift bellow the 'df1' and data column should be same
data_concat                     
```

```python
data_row_reindex=pd.concat([df,df1],ignore_index=True)
data_row_reindex
```

```python
data_append=df.append(df1) # it works same like concat but it is slow than concat
data_append
```

# difference between concatenate and append in pandas

Append function will add rows of second data frame to first dataframe iteraticely one by one. concat will do a single operation to finish the job, which makes it faster than append().

# Merge

```python
data_row=data_append
data_row
```

```python
data_merge_col=pd.merge(data_row,df2,on='id')
data_merge_col
```

# Joins

```python
df
```

```python
df1
```

Outer_Join

```python
data_outer=pd.merge(df,df1,on='id',how='outer')
data_outer
```

Inner_Join

```python
data_inner=pd.merge(df,df1,on='id',how='inner')
data_inner
```

Right_Join

```python
data_Right=pd.merge(df,df1,on='id',how='right')
data_Right
```

Left_Join

```python
data_Left=pd.merge(df,df1,on='id',how='left')
data_Left
```

# Melt

```python
dummy_data = {
            'A':{0:'a',1:'b',2:'c'},
            'B':{0:1,1:3,2:5},
            'C':{0:2,1:4,2:6}
}
df=pd.DataFrame(dummy_data)
```

```python
df
```

```python
pd.melt(df,id_vars=['A'],value_vars=['B'])
```

```python
pd.melt(df,id_vars=['A'],value_vars=['C'])
```
